```python
import pandas as pd
```

```python
df=pd.read_csv('weather_report.csv')
```

```python
def f(x):return None if x=='None' else x 
```

```python
l=df.columns
```

```python
for i in df.columns:df[i]=df[i].apply(f)
```

```python
x=df[df[l[0]].notnull()]
```

```python
index_column = x.index.tolist()
```

```python
n=[i for i in index_column if i%2==0]
```

```python
try:
    l=[df['PM2.5'].iloc[i] for i in n]
except IndexError:
    print(i)
```

```python
l=[float(i) for i in l]
l,len(l)
```

```python
x_new=pd.DataFrame(l,n,['PM2.5'])
x_new
```

```python
x_i=pd.DataFrame(x['PM2.5'].replace(l,None),df.index,['PM2.5'])
```

```python
x_i
```

```python
def f1(x):return None if x==None else float(x) 
```

```python
x_i['PM2.5']=x_i['PM2.5'].apply(f1)
```

```python
x_i.interpolate(limit_direction='both',inplace=True)
```

```python
x_i
```

```python
r=pd.DataFrame(x_i[x_i.index%2==0]['PM2.5'],n,['PM2.5'])
```

```python
r,x_new
```

```python
from sklearn.metrics import mean_squared_error
```

```python
rmse = mean_squared_error(x_new['PM2.5'],r['PM2.5'], squared = False)
```
